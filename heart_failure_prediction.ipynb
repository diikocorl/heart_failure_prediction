```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
```

```python
df.head(5)
```

```python
df.drop(['diabetes'],axis=1)
```

```python
data = df.values
X = data[:, 2:4]
y = data[:, -1].reshape(-1, 1) 
```

```python
print(X)
```

```python
## sử dụng thư viện MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
Xscaler = MinMaxScaler()
Xscaler.fit(X)

X = Xscaler.fit_transform(X)  ## Scale dữ liệu của x
yscaler = MinMaxScaler()
yscaler.fit(y)
y = yscaler.fit_transform(y)  ## Scale dữ liệu của y
```

```python
print('X: ',X[0:5])
print('\nY: ',y[0:5])
```

```python
print(X.shape)
```

```python
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)


### test 20%, train 80%
y_train = y_train.ravel()
y_test = y_test.ravel()

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report

from sklearn.neighbors import  KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

#KNN
KNN = KNeighborsClassifier(n_neighbors = 3)
KNN.fit(X_train, y_train)
y_predict_KNN = KNN.predict(X_test)
score_KNN = metrics.accuracy_score(y_predict_KNN,y_test)
print("Accuracy of KNN: ",score_KNN)
print(classification_report(y_test, y_predict_KNN))
print('------------------------------------------------------------')

#SVM
svm = SVC(kernel='linear')
svm.fit(X_train,y_train)
y_predict_svm = svm.predict(X_test)
score_svm = metrics.accuracy_score(y_predict_svm,y_test)
print("Accuracy of SVM: ",score_svm)
print(classification_report(y_test,y_predict_svm))
print('------------------------------------------------------------')

#NB
NB = GaussianNB()
NB.fit(X_train,y_train)
y_predict_NB = NB.predict(X_test)
score_NB = metrics.accuracy_score(y_predict_NB,y_test)
print("Accuracy of NB: ", score_NB)
print(classification_report(y_test,y_predict_NB))
print('------------------------------------------------------------')

#MLPC
MLPC = MLPClassifier(hidden_layer_sizes=(20, 20, 20))
MLPC.fit(X_train, y_train)
y_predict_MLPC = MLPC.predict(X_test)
score_MLPC = metrics.accuracy_score(y_predict_MLPC,y_test)
print("Accuracy of MLPC: ", score_MLPC)
print(classification_report(y_test,y_predict_MLPC))
```
